##Bring in libraries and dataframes and set indexes

```python
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Read in data
AllData=pd.read_csv('K:\Rainshelter\FARPotatoes\SpudsTDR_Table2.dat', #specify file path for data to read in
                         parse_dates=True, #tell the function to parse date columns to datetime formats
                         dayfirst=True, #tell the function that the day is before the year in the data i.e format='%d/%m/%Y %H:%M'
                         skiprows = [0,2,3], #leave out rows 1, 3 and 4 which have redundant information
                         index_col = 0, #Use the first column, which is Date, as an index
                         na_values = 'NAN')

#Bring in index data
AllDataIndex=pd.read_csv('C:\GitHubRepos\FarPotatoes\TDRIndex.csv',
                         index_col = 0)
#Apply indexes to data
AllDataTransposed = AllData.transpose()
AllDataIndexed = pd.concat([AllDataIndex,AllDataTransposed], axis=1)
AllDataIndexed.index.name='ColumnHeader'
AllDataIndexed.set_index(['Treatment','Position','Depth','TDR','MUX','MUX channel','Plot'], 
                        append=False, inplace=True)
AllDataIndex.sort_index()
Data=AllDataIndexed.transpose()
Data.sort(inplace=True)

#Calculate treatment means
DataMeans = Data.groupby(level=['Treatment','Position'],axis=1).mean()
```

## Set time slice to graph

```python
from datetime import timedelta #Import function needed for doing date deltas 
EndDate = Data.index.max(axis=0)  #Set end of series as most recent date in file
EndDateString = EndDate.strftime("%Y-%m-%d")  #Turn this to a string
PlotDuration = 3                              #Set the duration that you want to graph for
StartDate = EndDate - timedelta(weeks=PlotDuration)       #Set start date the specified duration before the end date 
StartDateString = StartDate.strftime("%Y-%m-%d")           #Turn that into a string
```

##Check temperature and volt data signals

```python
DataMeans.BattV_Avg.plot(figsize=(18,8))
DataMeans.PTemp_C_Avg.plot(figsize=(18,8))
```

```python
#Recalculate treatment means with indexes in different order
DataMeans = Data.groupby(level=['Position','Treatment'],axis=1).mean()
```

##Calculate and graph profile water content

```python
#Calculate the water content of the soil profile by multiplying the volumetric water content by each layers
#depth and summing.  The 0-15 layers are divided by 2 to average the two readings
ProfileWater = DataMeans.ix[:,'BetweenPlants'] * 250/2 + \
               DataMeans.ix[:,'NearPlant'] * 250/2 + \
               DataMeans.ix[:,'Furrow'] * 300 + \
               DataMeans.ix[:,'UnderRidge'] * 300 
#ProfileWater.ix[StartDateString:,:].plot(figsize=(18,18))
ProfileWater.plot(figsize=(18,18))
plt.ylabel('Profile Water (mm)', fontsize=30);
plt.tick_params(labelsize=25)
plt.title('Far Potato irrigation trial', fontsize=28);
```

##Graph treatment means by measurement position

```python
MeanLookSlice = DataMeans.ix[StartDateString:]

Fig = plt.figure(figsize=(18, 60))

def MakePlot(Position, Depth, LayerThickness, Ymax):
        Fig.add_subplot(8,1,Position)
        plt.title(Depth + '  VolumetricWaterContent', fontsize=28);
        plt.plot(MeanLookSlice.index,MeanLookSlice.ix[StartDateString:,Depth]*LayerThickness);
        plt.ylabel('mm', fontsize=22);
        plt.tick_params(labelsize=16);
        plt.ylim(10,Ymax);
        return;

MakePlot(1,'BetweenPlants',150,70)
MakePlot(2,'NearPlant',150,70)
MakePlot(3,'Furrow',300,100)
MakePlot(4,'UnderRidge',300,100)
```
